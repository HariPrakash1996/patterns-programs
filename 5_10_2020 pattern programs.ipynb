## pattern programs

```python
n=int(input("Enter n value :"))
for i in range(n) :
        print("*")
```

```python
n=int(input("Enter n value :"))
for i in range(n) :
        print("*",end=" ")
```

```python
n=int(input("Enter number of rows :"))
for i in range(n) :
        print("* "*n)
```
